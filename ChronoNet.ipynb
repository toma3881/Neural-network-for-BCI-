```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1234,stratify=y)
```

```python
X_train,X_val,y_train,y_val=train_test_split(X_train,y_train,test_size=0.25,random_state=1234,stratify=y_train)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Flatten,Dense, Dropout,Conv1D,MaxPooling1D,LeakyReLU,BatchNormalization,GlobalAveragePooling1D,AveragePooling1D
```

```python
from tensorflow.keras.layers import Input,Dense,concatenate,Flatten,GRU,Conv1D
from tensorflow.keras.models import Model
```

```python
X_train.shape
```

```python
def block(input):
  conv1 = Conv1D(32, 2, strides=2,activation='LeakyReLU',padding="same")(input)
  conv2 = Conv1D(32, 4, strides=2,activation='LeakyReLU',padding="causal")(input)
  conv3 = Conv1D(32, 8, strides=2,activation='LeakyReLU',padding="causal")(input)
  x = concatenate([conv1,conv2,conv3],axis=2)
  return x
```

```python
input= Input(shape=(77,32))
block1=block(input)
block2=block(block1)
block3=block(block2)
```

```python
gru_out1 = GRU(32,activation='tanh',return_sequences=True)(block3)
gru_out2 = GRU(32,activation='tanh',return_sequences=True)(gru_out1)
gru_out = concatenate([gru_out1,gru_out2],axis=2)
gru_out3 = GRU(32,activation='tanh',return_sequences=True)(gru_out)
gru_out = concatenate([gru_out1,gru_out2,gru_out3])
gru_out4 = GRU(32,activation='tanh')(gru_out)
```

```python
drop=block4=Dropout(0.5)(gru_out4)
predictions = Dense(1,activation='tanh')(drop)
model = Model(inputs=input, outputs=predictions)

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
scaler=StandardScaler()
```

```python
X_test=X_test.reshape(X_test.shape[0],-1)
```

```python
X_train=X_train.reshape(X_train.shape[0],-1)
```

```python
X_val=X_val.reshape(X_val.shape[0],-1)
```

```python
X_train=scaler.fit_transform(X_train)
X_val=scaler.fit_transform(X_val)
X_test=scaler.fit_transform(X_test)
```

```python
X_train.shape
```

```python
X_train=X_train.reshape(X_train.shape[0],77,32)
X_val=X_val.reshape(X_val.shape[0],77,32)
X_test=X_test.reshape(X_test.shape[0],77,32)
```

```python
from keras.callbacks import EarlyStopping
es=EarlyStopping(monitor='loss',mode='min',verbose=1,patience=2)
```

```python
model.fit(X_train, y_train,validation_data=(X_val,y_val), epochs=50,batch_size=64,callbacks=[es])
```
